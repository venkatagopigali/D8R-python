```python
t=(1,2,3,4)
t
```

2-d array using  tuple

```python
t=((1,2,3),(4,5,6),(0,0,0))
```

```python
a=np.array(t)
print(a)
print(a.ndim)
```

3-d using tuple

```python
t=(((1,2,3),(4,5,6)),((9,8,7),(6,5,4)))
```
