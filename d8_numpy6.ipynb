```python
import numpy as np
```

```python
a=np.arange(101,121)
```

```python
print(a)
```

```python
b=a.reshape(4,5)
```

```python
print(b)
```

```python
print(b[1,2])
```

```python
print(b[1:3:1,1:4:1])
```

```python
print(b[::3,::4])
```

```python
print(b[::,::2])
```

```python
print(b[::2,::2])
```

3-d slicing

```python
a=np.arange(101,137)
print(a)
```

```python
b=a.reshape(3,4,3)
```

```python
print(b)
```

```python
print(b[1,2,1])
```

```python
print(b[::,1:3:1,1:2:1])
```

```python
print(b[2::,1:3:1,0:1:1])
```

```python
print(b[0:3:1,0:4:3,0:3:2])
```

```python
print(b[0:3:2,1:4:2,0:3:2])
```

mathamatical functions

```python
a=np.array([1,2,1,3,4,5,5])
print(a)
```

```python
print(np.sum(a))
```

```python
print(np.max(a))
```

```python
print(np.min(a))
```

```python
print(np.cumsum(a).sum())
```

```python
print(np.prod(a))
```

```python
print(np.cumprod(a))
```

```python
print(np.cumprod(a).prod())
```

```python
print(np.cumprod(a).sum())
```

```python
print(np.std(a))
```

```python
print(np.var(a))
```

```python
print(np.mean(a))
```
