```python
import numpy as np
```

```python
l=[1,2,3,4]
l
print(type(l))
```

```python
a=np.array(l)
print(a)
print(type(a))
print(a.ndim)
print(a.shape)
print(a.size)
print(a.dtype)
print(a.itemsize)
print(a.nbytes)   # size*itemsize
```

2-d array

```python
l=[[1,2,3],[4,5,6.0]]
l
```

```python
a=np.array(l)
print(a)
print(a.ndim)
print(a.shape)
print(a.size)
print(a.dtype)
print(a.itemsize)
print(a.nbytes)
```

3-d array'

```python
l=[[[1,2,3],[4,5,6],[7,8,9]],[[8,9,0],[3,4,5],[6,7,'8']]]
l
```

```python
a=np.array(l)
print(a)
print(a.ndim)
print(a.shape)
print(a.size)
print(a.dtype)
print(a.itemsize)
print(a.nbytes)
```
